```python
import pandas as pd
```

```python
data = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv', sep=',')
data = data.apply(pd.to_numeric, errors='coerce')
data = data.astype(int)

data
```

```python

range_col_names = ['BMI', 'MentHlth', 'PhysHlth']
quantiles_values = [[0.2, 0.4, 0.6, 0.8],[0.5], [0.5]]



for index in range(len(range_col_names)): 
    quantiles = data[range_col_names[index]].quantile(quantiles_values[index])
    ranges = [data[range_col_names[index]].min()] + quantiles.tolist() + [data[range_col_names[index]].max()]
    range_labels = [f'{int(ranges[i])}-{int(ranges[i+1])}' for i in range(len(ranges) - 1)]

    for i, range_label in enumerate(range_labels):
        min_val = ranges[i]
        max_val = ranges[i + 1]
        data[f'{range_col_names[index]}_{range_label}'] = ((data[range_col_names[index]] >= min_val) & (data[range_col_names[index]] < max_val)).astype(int)

    data.drop(columns=[range_col_names[index]], inplace=True)


data
```

```python
dumie_col_names = ['GenHlth', 'Age', 'Education', 'Income']

for col in dumie_col_names:
    dummies = pd.get_dummies(data[col], prefix=f'{col}_')
    dummies = dummies.astype(int)
    data = pd.concat([data, dummies], axis=1)
    data.drop(columns=[col], inplace=True)

data
```

```python
data.to_csv('output.csv', index=False)
```
